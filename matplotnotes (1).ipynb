```python
#                   MATPLOTLIB
#this is used for plotting graphs and get data from graphs from the given data.

import matplotlib.pyplot as plt  #we use "pyplot" in matplotlib.

plt.plot([2,4,6,4],color='g')# this internally plots the graph
                             # when only one list or array is provided it plots the graph against the list's indices.
                             # instead of "color='g'" 'g' can be used.note:this is if we are just specifing the color but not color with shape like "g^" which means green triangle.
                             # note: 'color','g' can't be used in "plot" but can be         
plt.ylabel("numbers")# y axis name
plt.xlabel("indices")# x axis name
plt.title("Myplot")# title/name of the graph.
plt.show()        #to print/show the graph.
```

```python
plt.plot((1,2,3,4),[1,4,9,16],"ro")#this plots graph against these 2 lists."ro" in this line prints a "red dot" at every point given instead of a blue line like in the previous graph 
plt.ylabel("squares")              # here we can't use "color='ro'" cuz 'ro' doesn't only mean "red" but "red dots".
plt.xlabel("numbers")
plt.grid()# for grid lines in the graph. grid lines improves redability.
plt.show()
```

```python
import numpy as np
t=np.arange(0.,5.,0.2)

# blue dashes "b--", red squares "rs", green traingles "g^"
plt.plot(t,t**2,"b--",label="^2")# label is for specifing the line's significance.
plt.plot(t,t**2.2,"rs",label="^2.2")
plt.plot(t,t**2.5,"g^",label="^2.5")
plt.grid()
plt.legend()# this prints all "labels" in a box on the top left corner in the graph.
plt.show()
```

```python
x=[1,2,3,4]
y=[1,4,9,16]
w=plt.plot(x,y,linewidth=5)#"linewidth" increases the width of the line.
                         # "linewidth" only works when the plotted part is in the form of a line not squares,dots,triangles etc.
plt.show()
```

```python
x1=[1,2,3,4]
y1=[1,4,9,16]
x2=[1,2,3,4]
y2=[2,4,6,8]
lines=plt.plot(x1,y1,x2,y2)

plt.setp(lines[0],color='r',linewidth=2,label='**2')#"plt.setp()"is used to edit the lines in the graph saperately and these are the 2 ways for doing this.
plt.setp(lines[1],'color','g','linewidth',2)
plt.grid()
plt.legend()
plt.show()
#****************************************************************
# NOTE:lines=plt.plot(x1,y1,'g',x2,y2,'r',linewidth=2) can also be used instead of "plt.setp()".
```

```python
import numpy as np

def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
t1=np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.02)

plt.figure(1)

plt.subplot(211)#"211" means 2 rows,1 column,1st row/block. ie, the 1st block of the 2 rows and 1 column
plt.grid()
plt.plot(t1,f(t1),'b-')

plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2),'r--')
plt.plot()
plt.show()# as "show" is used we can't edit or add new plots to the existing graphs by using "plt.figure()" again but it re prints the same graphs with the new plots eg: given below.


plt.figure(1)
plt.subplot(211)
plt.plot(t1,'b-')
plt.show()
```
