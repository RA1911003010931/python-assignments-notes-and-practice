```python
#            concatenating dataframes
import pandas as pd
india_weather=pd.DataFrame({"city":['mumbai','delhi','banglore'],
                           "temperature":[32,45,30],
                            "humidity":[80,60,78]
                           })# this is called "hashmap" it looks like a dictionary.
# THIS IS AN OTHER WAY OF CREATING A DATAFRAME 
# "hasmap" could be given a variable like w={hashap}
india_weather
```

```python
us_weather=pd.DataFrame({"city":['ny','chicago','orlando'],
                           "temperature":[21,14,35],
                            "humidity":[88,65,75]
                           ,})
us_weather
```

```python
df=pd.concat([india_weather,us_weather])# this combines/concatenates both dataframes to a single dataframe
df                                      # but the indices doesn't change
```

```python
# for ignoring pre existing indices
df=pd.concat([india_weather,us_weather],ignore_index=True)
df                     #now properly ordered indices are printed.
```

```python
df=pd.concat([india_weather,us_weather],axis=1)# this is for concatenating sideways.
df
```

```python
# merging dataframes
w={'city':['mumbai','delhi','banglore','hyderabad'],'temperature':[32,45,30,40]}
temp_df=pd.DataFrame(w)
temp_df
```

```python
humidity_df=pd.DataFrame({'city':['mumbai','delhi','banglore'],'humidity':[68,65,75]})
humidity_df
```

```python
df=pd.merge(temp_df,humidity_df,on='city')#"on='city'" is used because city is a common column in both dataframes
df                                        # by default python mearges the dataframes(without using "on=city") based on what column is common in them."on=xyz" is useful when there r more than 1 columns in common.

# note: "hyderabad" in one of the dataframes is missing in the mearged dataframe cuz hyd isn't there in one dataframe.
```

```python
df=pd.merge(temp_df,humidity_df,on='city',how='outer')# "how='outer'" helps us print the uncommon row(hyd in this case as hyd isnt in one dataframe and the mearging is "on='city')")
df
```
